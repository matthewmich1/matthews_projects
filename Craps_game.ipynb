```
def rules():
  print('Craps Rules:')
  print('- A dice will roll to see who starts first.')
  print('- Each player first buys an initial amount of chips where each chip is worth $60.')
  print('- Each player gets to select 1 type of craps bet to play where they place a bet with some of their chips in hope of winning extra chips.')
  print("- You can't bet with more chips than you have at your disposal")
  print('- The game ends when a player loses all their chips or a player decides to leave.')
  print('- Pass line: roll a seven or 11 and you win. Roll a 2,3,12 and you lose. Any other number, the dice keeps rolling until you get that number. If it appears before a 7, you win.')
  print("- Don't pass line: roll a 2,3 and you win. Roll a 12 and you keep your chips. Any other number, the dice keeps rolling until you get that number. If a 7 appears first, you win.")
  print('- Come bet: roll a 7 or 11 and you win.')
  print('- Place bet: minimum of $6 for 6 or 8 roll and minimum of $5 for 4,5,9,10. This are the numbers you can bet for the Place Bet. Roll the number you bet on and you win.')
  print('- Field bet: Bet for 2,3,4,9,10,11 or 12. Roll a 2 or 12 and you win double your bet.')
  print('- Proposition bet: bet that both dice will show a 2,3,4,5. You can win big on this bet.')
  print('- Horn bet: bet for 2,3,11,12.')
  print('- C and E bet: You can choose a C bet which is a bet on any roll, an E bet which is a bet on 11, or both.')
  print('- Any seven bet: bet on a 7.')
```

```
#decides which player has first bet
import random
def first_turn(player1, player2):
  roll = random.randint(0,1)
  if roll == 0:
    turn = player1
    print(player1,'goes first.')
  else:
    turn = player2
    print(player2,'goes first.')
```

```
def pass_line(bet, player_chips,player):
  while bet > player_chips or bet < 0:
    bet = int(input('This bet is greater than your total number of chips or less than 0. Place bet: '))

  dice_roll = random.randint(1,6) + random.randint(1,6)

  if dice_roll in (7,11):
    print('You rolled a',dice_roll,'and won 4 extra chips!')
    player_chips += (bet + 4)
    return player_chips

  elif dice_roll in (2,3,12):
    print('You rolled a',dice_roll,'and lost your bet.')
    player_chips -= bet
    return player_chips

  else:  # point numbers
    num_rolled = dice_roll
    dice_roll = random.randint(1,6) + random.randint(1,6)
    print('You rolled a',dice_roll)

    while dice_roll != num_rolled:
      dice_roll = random.randint(1,6) + random.randint(1,6)

      if dice_roll == num_rolled:
        print(player,'rolled a',dice_roll,'before a 7 so you win an extra chip.')
        player_chips += (bet + 1)
        return player_chips

      if dice_roll == 7:
        print(player,'rolled a 7 so you lose your bet.')
        player_chips -= bet
        return player_chips
```

```
def do_not_pass_bar(bet,player_chips,player): #bet for 2 or 3
  while bet > player_chips or bet < 0:
    bet = int(input('This bet is greater than your total number of chips or less than 0. Place bet: '))

  dice_roll = random.randint(1,6) + random.randint(1,6)

  if dice_roll == 2 or dice_roll == 3:
    print(player,'rolled a',dice_roll)
    player_chips += (bet + 3)
  elif dice_roll == 12:
    print(player,'rolled a',dice_roll)
    print('You keep your chips.')
  else:
    num_rolled = dice_roll
    dice_roll = random.randint(1,6) + random.randint(1,6)

    while dice_roll != num_rolled:
      dice_roll = random.randint(1,6) + random.randint(1,6)

      if dice_roll == num_rolled:
        print(player,'rolled a',dice_roll)
        print('You lose your bet.')
        player_chips -= bet
        return player_chips

      if dice_roll == 7:
        print(player,'rolled a',dice_roll)
        print('You win your bet and 2 extra chips.')
        player_chips += (bet + 2)
        return player_chips
```

```
def field_bet(bet,player_chips,player):
  while bet > player_chips or bet < 0:
    bet = int(input('This bet is greater than your total number of chips or less than 0. Place bet: '))

  dice_roll = random.randint(1,6) + random.randint(1,6)

  if dice_roll in (3,4,9,10,11):
    print(player,'rolled a',dice_roll,'so you win 2 extra chips.')
    player_chips += (bet + 2)

  elif dice_roll in (2,11):
    print(player,'rolled a',dice_roll,'so you win double your bet!')
    player_chips += (bet*2)

  else:
    print(player,'rolled a',dice_roll,'so you lose your bet.')
    player_chips -= bet
```

```
def proposition_bet(bet,player_chips,player):
  while bet > player_chips or bet < 0:
    bet = int(input('This bet is greater than your total number of chips or less than 0. Place bet: '))

  dice1 = random.randint(1,6)
  dice2 = random.randint(1,6)

  if dice1 in (2,5) and dice2 == dice1:
    print(player,'got a double!',dice1,'and',dice2)
    player_chips += (bet * 10)

  elif dice1 in (3,4) and dice2 == dice1:
    print(player,'got a double!',dice1,'and',dice2)
    player_chips += (bet * 9)

  else:
    print(player,'got',dice1,'and',dice2,'so you lose.')
    player_chips -= (bet * 2)
```
