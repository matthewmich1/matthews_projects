```python
def quad(a,b,c):
  import numpy as np
  import math
  
  #for real solution to exist, z must be greater or equal to 0
  z = b ** 2 - 4*a*c
  surd = np.sqrt(z)
  if z < 0:
    print('No real solution exist')
  else:
    print('The real solution x1 and x2 are:')
    r1,r2 = (-b + surd)/(2*a), (-b - surd)/(2*a)
    a1 = np.round(r1,4)
    a2 = np.round(r2,4)

    return [a1, a2]
quad(1,3,-7)
```
